````python
# Query by dates
## find all the 1900 competitions

df.query("Year == 1900")
````

````python
## That found the ones in the year 1900 ONLY and not the whole century
## return all the comps between 1940 and 1960

df.query("1940 <= Year <= 1960")

````

````python
## Now we want to find the Black athletes during this time

df.query("Year == 1936 & NOC == 'USA'")
````

````python
df.query("NOC == 'IND' and Medal == 'Gold'")

## can use & or and 
## we are looking for India when they won gold
## You have to put strings (words) in ' ' 
````

````python
## Now we are looking for gold OR silver

india_df = df.query("NOC == 'IND'")

india_df
````

````python
india_df.query("Medal == 'Gold' or Medal == 'Silver' or Medal == 'Bronze'")
````

````python
## WOOOOOOO that worked!!!

india_df.query("Medal == 'Gold' or Medal == 'Silver' or Medal == 'Bronze'").sort_values(by="Medal")
````

## create more cells as necessary

````python
## bring in our diamonds dataset
diamonds_df = pd.read_csv("https://raw.githubusercontent.com/sandeepmj/datasets/refs/heads/main/diamonds.csv")
diamonds_df
````

````python
diamonds_df["diamond cut"].nunique()

##Number of unique cuts
````

````python
## What is the break down diamond_cut

diamonds_df["diamond cut"].value_counts(ascending = True)
````

````python
diamonds_df["diamond cut"].value_counts(ascending = False).to_frame("Quantity")
````

````python
## What we just did above is called Daisy chaining -- > it does all the tasks. 
## The to_frame can be added to almost anything - he suggests using it often especially when sharing work with editors
````
